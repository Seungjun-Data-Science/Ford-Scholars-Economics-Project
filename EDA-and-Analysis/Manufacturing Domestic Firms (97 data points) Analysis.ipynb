```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mp
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('manuf_domest.csv')
```

```python
del df['Unnamed: 0']
```

```python
df.info()
```

```python
# Unique Number of Manufacturing Firms for analysis
df.cik.nunique()
```

## Creating new variables [field]/Assets

```python
# Creating new variable (total income / Assets)
df['Total_Income_scaled']=\
df['GrossProfit']/df['Assets']
```

```python
# Creating new variable (Liabilties / Assets)
df['Liabilities_scaled']=\
df['Liabilities']/df['Assets']
```

```python
# Creating new variable (R&D / Assets)
df['ResearchAndDevelopmentExpense_scaled']=\
df['ResearchAndDevelopmentExpense']/df['Assets']
```

```python
# Creating new variable (R&D / Assets)
df['PropertyPlantAndEquipmentNet_scaled']=\
df['PropertyPlantAndEquipmentNet']/df['Assets']
```

**Total Income**

```python
# Mean of (Total Income/ Assets) of firms in Manufacturing per year
total_income_mean_scaled =\
pd.DataFrame(df[df['Total_Income_scaled'].notnull()].\
             loc[:,['cik','fy','Type of Firm','Total_Income_scaled']].\
             groupby(['fy','Type of Firm'])['Total_Income_scaled'].mean()).reset_index()
```

```python
# Total Income normalized by Assets of MNCs over time
sns.pointplot(x=total_income_mean_scaled.fy, y=total_income_mean_scaled.Total_Income_scaled)

plt.title('Mean of Total Income normalized by Assets of Domestic Firms over time (Manufacturing)')
plt.ylabel('Mean of Total Income normalized')
```

**R&D**

```python
# Mean of (R & D / Assets) of firms in industry 2 per year
r_and_d_mean_scaled =\
pd.DataFrame(df[df['ResearchAndDevelopmentExpense_scaled'].notnull()].\
             loc[:,['cik','fy','Type of Firm','ResearchAndDevelopmentExpense_scaled']].\
             groupby(['fy','Type of Firm'])['ResearchAndDevelopmentExpense_scaled'].mean()).reset_index()
```

```python
# R & D normalized by Assets of MNCs over time
sns.pointplot(x=r_and_d_mean_scaled.fy,
         y=r_and_d_mean_scaled.ResearchAndDevelopmentExpense_scaled)

plt.title('Mean of R&D Expense normalized by Assets of Domestic Firms over time (Manufacturing)')
plt.ylabel('Mean of R&D Expense normalized')
```

**Liabilities**

```python
# Mean of (Liabilities/ Assets) of firms in Manufacturing per year
liabilities_mean_scaled =\
pd.DataFrame(df[df['Liabilities_scaled'].notnull()].\
             loc[:,['cik','fy','Type of Firm','Liabilities_scaled']].\
             groupby(['fy','Type of Firm'])['Liabilities_scaled'].mean()).reset_index()
```

```python
# Liabilities normalized by Assets of MNCs over time
sns.pointplot(x=liabilities_mean_scaled.fy,
y=liabilities_mean_scaled.Liabilities_scaled)

plt.title('Mean of Liabilities normalized by Assets of Domestic Firms over time (Manufacturing)')
plt.ylabel('Mean of Liabilities normalized')
```

#### Net Capital

```python
# Mean of (Net Capital/ Assets) of firms in Manufacturing per year
net_capital_mean_scaled =\
pd.DataFrame(df[df['PropertyPlantAndEquipmentNet_scaled'].notnull()].\
             loc[:,['cik','fy','Type of Firm','PropertyPlantAndEquipmentNet_scaled']].\
             groupby(['fy','Type of Firm'])['PropertyPlantAndEquipmentNet_scaled'].mean()).reset_index()
```

```python
# Liabilities normalized by Assets of MNCs over time
sns.pointplot(x=net_capital_mean_scaled.fy,
y=net_capital_mean_scaled.PropertyPlantAndEquipmentNet_scaled)

plt.title('Mean of Net Capital normalized by Assets of Domestic Firms over time (Manufacturing)')
plt.ylabel('Mean of Net Capital normalized')
```
