```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('josh_final_dataset_7.12.csv')
```

```python
del df['Unnamed: 0']
```

```python
net_income_loss = pd.read_csv('net-income-loss_clean.csv')
gross_profit = pd.read_csv('gross-profit_clean.csv')
profit_loss = pd.read_csv('profit-loss_clean.csv')
# total_income = pd.read_csv('total-income_clean.csv')
```

```python
del net_income_loss['Unnamed: 0']
del gross_profit['Unnamed: 0']
del profit_loss['Unnamed: 0']
# del total_income['Unnamed: 0']
```

```python
df = df.merge(net_income_loss.loc[:,['cik','fy','NetIncomeLoss']], on=['cik','fy'],how='outer')
df = df.merge(gross_profit.loc[:,['cik','fy','GrossProfit']], on=['cik','fy'],how='outer')
df = df.merge(profit_loss.loc[:,['cik','fy','ProfitLoss']], on=['cik','fy'],how='outer')
```

## Checking for STILL REMAINING duplicates

```python
df[df.duplicated(['cik','fy'],keep=False)]
```

```python
df.drop_duplicates(['cik','fy'],keep='last',inplace=True)
```

## Saving the file
